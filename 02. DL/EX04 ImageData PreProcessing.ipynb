```python
from zipfile import ZipFile # zip 파일을 다룰 때 사용
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image # 사진 데이터를 다룰 때 이용
```

```python
# 압축파일 해제
zip_file_path = './data/cats_and_dogs_filtered.zip'

with ZipFile(zip_file_path) as zf:
  zf.extractall('./data')
```

```python
train_cats_dir = './data/cats_and_dogs_filtered/train/cats'
train_dogs_dir = './data/cats_and_dogs_filtered/train/dogs'
test_cats_dir = './data/cats_and_dogs_filtered/test/cats'
test_dogs_dir = './data/cats_and_dogs_filtered/test/dogs'
```

```python
train_cats_fn = os.listdir(train_cats_dir)
train_dogs_fn = os.listdir(train_dogs_dir)
test_cats_fn = os.listdir(test_cats_dir)
test_dogs_fn = os.listdir(test_dogs_dir)
```

```python
len(train_cats_fn), len(train_dogs_fn), len(test_cats_fn), len(test_dogs_fn)
```

```python
from tqdm.auto import tqdm
```

```python
def prepro_img(dir_path, file_names, img_size = (224, 224)):
  imgs = []
  for fn in tqdm(file_names):

    path = os.path.join(dir_path, fn)

    img = Image.open(path).resize(img_size)
    imgs.append(np.array(img))

  return np.array(imgs)
```

```python
X_train_cats = prepro_img(train_cats_dir, train_cats_fn)
X_train_dogs = prepro_img(train_dogs_dir, train_dogs_fn)
X_test_cats = prepro_img(test_cats_dir, test_cats_fn)
X_test_dogs = prepro_img(test_dogs_dir, test_dogs_fn)

print('훈련 데이터 : ', X_train_cats.shape, X_train_dogs.shape)
print('테스트 데이터 : ', X_test_cats.shape, X_test_dogs.shape)
```

```python
X_train = np.concatenate((X_train_cats, X_train_dogs), axis = 0)
X_test = np.concatenate((X_test_cats, X_test_dogs), axis = 0)
```

```python
X_train.shape, X_test.shape
```

```python
y_train = np.array([0] * 1000 + [1] * 1000)
y_test = np.array([0] * 500 + [1] * 500)

y_train.shape, y_test.shape
```

```python
np.savez_compressed(
    './data/np_cat_vs_dog.npz',
    X_train = X_train,
    X_test = X_test,
    y_train = y_train,
    y_test = y_test
)
```
