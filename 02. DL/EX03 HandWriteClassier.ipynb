```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.datasets import mnist
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_test_new = X_test.reshape(-1, 784)
X_train_new = X_train.reshape(-1, 784)

X_train = X_train_new / 255.0
X_test = X_test_new / 255.0
```

```python
X_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, InputLayer
```

```python
model = Sequential()
model.add(InputLayer(shape = (28, 28)))
model.add(Flatten())
model.add(Dense(32, activation = 'softmax'))
model.add(Dense(64, activation = 'softmax'))
model.add(Dense(32, activation = 'softmax'))
model.add(Dense(10, activation = 'softmax'))
```

```python
model = Sequential()
model.add(InputLayer(shape = (28, 28)))
model.add(Flatten())
model.add(Dense(32, activation = 'relu'))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(32, activation = 'relu'))
model.add(Dense(10, activation = 'softmax'))
```

```python
model.compile(
    loss = 'sparse_categorical_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy']
)
```

```python
h = model.fit(X_train, y_train, validation split = 0.2, epochs = 20)
```

```python
plt.figure(figsize = (10, 3))
plt.plot(h.history['accuracy'], label = 'acc')
plt.plot(h.history['val_accuracy'], label = 'val_acc')
plt.legend()
plt.show()

# 약 7번째 epoch 이후에 train acc는 점점 높아지는데 비해, val acc는 비교적 성능 개선이 이루어지지 않음
# -> 과대적합 의심
# 과대적합을 해소하기 위해 콜백함수(조기학습중단 & 모델중간저장)를 통해 완성시켜보자
```

### 1. optimizer 종류
- Batch GD : 전체 데이터를 사용해서 학습
- SGD : 일부 데이터르 사용하여 학습
- Momentum : 관성, 이전의 기술기 값을 적용시켜 학습
- AdaGrad : 보폭의 크기를 줄여나가면서 안정성 위주로 학습
- Adam : Momentum + AdaGrad(관성, 보폭의 크기를 모두 고려)
- 현재는 Adam을 가장 많이 사용하는 추세, 최적화 함수로서 성능이 매우 높은 편
